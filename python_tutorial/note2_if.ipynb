```python
volt = 78

if volt >= 90:
    ch_Led = "Red"
    status = "Active"
elif volt >= 80:
    ch_Led = "Orange"
    status = "Active"
elif volt >= 70:
    ch_Led = "Yellow"
    status = "Active"
elif volt >= 60:
    ch_Led = "Green"
    status = "Active"
else:
    ch_Led = "None"
    status = "Under_Threshold" # 여기가 바로 '기준치 미달' 상태

# 출력부 (Print)
if ch_Led != "None":
    print(f"{volt}V 인가 시 {ch_Led} 램프 점등 (상태: {status})")
else:
    print(f"{volt}V 인가 시 램프 미점등 (상태: {status})")
```
